```
data = pd.read_csv("automobile.csv", sep = ',')
data
```

```
mean = data.mean()
mean
```

```
mean_data = ["normalized_losses", "wheel_base", "length","height","curb_weight","engine_size","bore","stroke"]
geometric_mean = stats.gmean(data.loc[:,mean_data])
geometric_mean
```

```
harmonic_mean = stats.hmean(data.loc[:,mean_data])
harmonic_mean
```
